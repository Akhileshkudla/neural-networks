```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.layers import Dense, InputLayer

import tensorflow as tf

from time import time

from mlxtend.plotting import plot_confusion_matrix

%matplotlib inline
```

```python
df_train = pd.read_csv("/data/MNIST/mnist_train.csv", header=None)
df_test = pd.read_csv("/data/MNIST/mnist_test.csv", header=None)
```

```python
df_train.shape, df_test.shape
```

```python
28 * 28 
```

```python
plt.imshow(df_train.iloc[0, 1:].reshape((28, 28)), cmap="gray")
plt.title(df_train.iloc[0, 0])
```

```python
X_train = df_train.iloc[:, 1:].values/255
y_train = df_train.iloc[:, 0].values
X_test = df_test.iloc[:, 1:].values/255
y_test = df_test.iloc[:, 0].values
```

```python
fig, _ = plt.subplots(10, 5, figsize = (8, 15))
for i, ax, in enumerate(fig.axes):
    ax.imshow(X_train[i].reshape(28, 28), cmap = "gray")
    ax.set_title(y_train[i])
plt.tight_layout()
```

```python
X_train[0]
```

```python
(pd.Series(y_train).value_counts()/len(y_train)).plot.bar()
```

```python
(pd.Series(y_test).value_counts()/len(y_test)).plot.bar()
```

```python
n_x = X_train.shape[1]

tf.set_random_seed(1)
np.random.seed(1)

tensor_board = keras.callbacks.TensorBoard(log_dir="/tmp/tf/logs/%d" % time())

model = keras.Sequential()
model.add(InputLayer((n_x, )))
model.add(Dense(units=10, activation="softmax"))

model.summary()

model.compile(loss=keras.losses.categorical_crossentropy
            , metrics=["accuracy"]
            , optimizer="adam")

model.fit(X_train, keras.utils.to_categorical(y_train)
          , batch_size=256
          , validation_data = (X_test, keras.utils.to_categorical(y_test))
          , epochs=10
          , verbose = 1
          , callbacks=[tensor_board])
```

```python
n_x = X_train.shape[1]

tf.set_random_seed(1)
np.random.seed(1)

tensor_board = keras.callbacks.TensorBoard(log_dir="/tmp/tf/logs/%d" % time())

model = keras.Sequential()
model.add(InputLayer((n_x, )))
model.add(Dense(units=100, activation="relu"))
model.add(Dense(units=10, activation="softmax"))

model.summary()

model.compile(loss=keras.losses.categorical_crossentropy
            , metrics=["accuracy"]
            , optimizer="adam")

model.fit(X_train, keras.utils.to_categorical(y_train)
          , batch_size=256
          , validation_data = (X_test, keras.utils.to_categorical(y_test))
          , epochs=10
          , verbose = 1
          , callbacks=[tensor_board])
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = model.predict_classes(X_test).flatten()
plot_confusion_matrix(confusion_matrix(y_test, y_test_pred), figsize=(6, 6))
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten
```

```python
# LeNet architecture
```

```python
n_x = X_train.shape[1]

tf.set_random_seed(1)
np.random.seed(1)

tensor_board = keras.callbacks.TensorBoard(log_dir="/tmp/tf/logs/%d" % time())

model = keras.Sequential()
model.add(InputLayer((28, 28, 1)))

model.add(Conv2D(32, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=100, activation="relu"))
model.add(Dense(units=10, activation="softmax"))

model.summary()

model.compile(loss=keras.losses.categorical_crossentropy
            , metrics=["accuracy"]
            , optimizer="adam")

model.fit(X_train.reshape(-1, 28, 28, 1)
          , keras.utils.to_categorical(y_train)
          , batch_size=256
          , validation_data = (X_test.reshape(-1, 28, 28, 1)
                               , keras.utils.to_categorical(y_test))
          , epochs=10
          , verbose = 1
          , callbacks=[tensor_board])
```
