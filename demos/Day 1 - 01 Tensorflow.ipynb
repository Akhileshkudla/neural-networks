```python
def is_outlier(y):
    q1, q3, = np.percentile(y, [25, 75])
    iqr = q3 - q1
    uw = min(np.max(y), q3 + 1.5 * iqr)
    lw = max(np.min(y), q1 - 1.5 * iqr)

    print("q1: %f q3: %f, iqr: %f, uw: %f lw: %f" %  (q1, q3, iqr, uw, lw))

    return np.where((y > uw) | (y < lw))[0]

outliers = is_outlier(y)
outliers.shape, y[outliers] 
```

# Tensorflow components


- placeholder (takes input to the graph)
- variables (that stores data during computation)
- optimizer (you can define a cost function as a function
of paramters (thetas) and optimizes (minimize/maximize) the 
value of cost function by finding the right set of values of the paramers. 
It finds the optimum value value by taking derivatives. 
How does it know about the derivatives? It knows the derivatives 
because in the compuation graphs allows only tensorflow recognized operation

- graph (you define a flow of data in the form of a computation graph)

```python
tf.reset_default_graph()
tf.set_random_seed(1)

n_x = X_train.shape[1]

X = tf.placeholder(dtype=tf.float32, shape=(None, n_x))
y = tf.placeholder(dtype = tf.float32, shape=(None,))

print("X", X)
print("y", y)

W = tf.Variable(tf.truncated_normal((n_x, 1), stddev=0.05))
b = tf.Variable(0.0)

print("W", W)
print("b", b)

Z = b + tf.matmul(X, W)
print("Z", Z)

y_pred = tf.reshape(tf.nn.relu(Z), (-1, ))

print("y_pred", y_pred)

error = y_pred - y

print("Error", error)

mse = tf.reduce_mean(tf.square(error))

print("mse", mse)

cost = tf.reduce_mean(tf.square(error) + 5.0 * tf.nn.l2_loss(W))

global_step = tf.Variable(0, dtype=tf.int32, trainable=False)

learning_rate = 0.01

opt = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost, global_step = global_step)

tf.summary.scalar("mse", mse)
tf.summary.histogram("W", W)
agg = tf.summary.merge_all()

writer = tf.summary.FileWriter(graph=tf.get_default_graph()
                            , logdir="/tmp/tf/logs/%d" % time())
saver = tf.train.Saver()
with tf.Session() as sess:
    scores = []
    
    weights = []
    
    sess.run(tf.global_variables_initializer())
    for i in range(500):
        mse_, _, W_, agg_, global_step_ = sess.run([mse, opt, W, agg, global_step]
                        , feed_dict={X: X_train, y: y_train})
        #print(i, mse_)
        scores.append(mse_)
        writer.add_summary(agg_, global_step_)
        weights.append(W_.ravel())
    tf.train.Saver()
    saver.save(sess=sess, save_path="/tmp/tf/chk")
        
    pd.Series(scores).plot()
    plt.title("Cost curve")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    
    y_test_pred = sess.run(y_pred, feed_dict = {X: X_test})
    test_mse = sess.run(mse, feed_dict = {X: X_test, y: y_test})
    print("test mse", test_mse)
```

```python
df = pd.read_csv("/data/kaggle/house-prices/data_combined_cleaned.csv")
df = df[~df.SalesPrice.isnull()]
X = pd.get_dummies(df.drop(columns=["Id", "SalesPrice"]), drop_first=True).values
X = (X - X.mean(axis=0))/ X.std(axis=0)
y = np.log(df.SalesPrice).values

indices = np.arange(len(X))

np.random.seed(1)
np.random.shuffle(indices)

X = X[indices]
y = y[indices]

training_size = int(len(X) * 0.7)

X_train, y_train = X[:training_size], y[:training_size]
X_test, y_test = X[training_size:], y[training_size:]

n_x = X_train.shape[1]
```

```python
# Hyper parameters
learning_rate = 0.01
epochs = 1000
reg_param = 5.0
chk_path = "/tmp/tf/chk"
```

```python
# Block for computation graph

tf.reset_default_graph()

regularizer = tf.contrib.layers.l2_regularizer(reg_param)
X = tf.placeholder(tf.float32, shape=(None, n_x))
y = tf.placeholder(tf.float32, shape=(None,1))
y_pred = tf.layers.dense(X, units=1, activation=tf.nn.relu, kernel_regularizer=regularizer, name="layer1")
mse = tf.losses.mean_squared_error(y, y_pred)

opt = tf.train.GradientDescentOptimizer(learning_rate).minimize(mse)

tf.summary.scalar("mse", mse)

merged = tf.summary.merge_all()

tf.global_variables()
```

```python
# This blocked is used in training time
writer = tf.summary.FileWriter("/tmp/tf/logs/%d" % time(), tf.get_default_graph())
saver = tf.train.Saver()
with tf.Session() as sess: 
    sess.run(tf.global_variables_initializer())
    values = {X: X_train, y: y_train.reshape(-1, 1)}
    for epoch in range(epochs):
        _, merged_, mse_ = sess.run([opt, merged, mse], feed_dict=values)
        writer.add_summary(merged_, epoch)
    print("Training mse: ", mse_)
    saver.save(sess, chk_path)
```

```python
# Loads the model from the disk and find the performance numbers fro
with tf.Session() as sess: 
    sess.run(tf.global_variables_initializer())
    saver.restore(sess, chk_path)
    values = {X: X_test, y: y_test.reshape(-1, 1)}
    mse_ = sess.run(mse, feed_dict=values)
    print(mse_)
```
